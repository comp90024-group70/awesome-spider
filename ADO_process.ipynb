```python
import json
import pandas as pd
```

```python
times_dic = []
sizes_dic = []
terms_dic = []
with open('../Data/ADO.json', 'r', encoding='utf-8') as f:
    data = json.load(f)
    for i in range(len(data)):
        for j in range(10):
            times, topics = iter(data[i].items())
            times_dic.append(times[1])
            size, terms = topics[1][j].items()
            sizes_dic.append(size[1])
            terms_dic.append(terms[1][1:15])
            
df_ADO = pd.DataFrame({'Time': times_dic, 'Size': sizes_dic, 'Terms': terms_dic})

def Extract(lst):
    return [item[0] for item in lst]
for i in range(len(df_ADO)):
    df_ADO['Terms'][i] = Extract(df_ADO['Terms'][i])

df_ADO['Terms'] = df_ADO['Terms'].map(str)
```

```python
df_ADO_topic = df_ADO[df_ADO['Terms'].str.contains('job|election|momey|covid|support|australia|family|country|governemnt|labor|vote')]

df_ADO_fam= df_ADO[df_ADO['Terms'].str.contains('family|home|house')]
df_ADO_fam= df_ADO_fam[df_ADO_fam['Terms'].str.contains('election|momey|support|australia|country|governemnt|vote')]
df_ADO_job = df_ADO[df_ADO['Terms'].str.contains('job|labor|work')]
df_ADO_job= df_ADO_job[df_ADO_job['Terms'].str.contains('election|momey|support|australia|country|governemnt|vote')]
df_ADO_covid = df_ADO[df_ADO['Terms'].str.contains('job|labor|work')]
df_ADO_covid = df_ADO_covid[df_ADO_covid['Terms'].str.contains('election|momey|support|australia|country|governemnt|vote')]

df_ADO_fam['Total'] = df_ADO_fam.groupby(['Time'])['Size'].transform('sum')
df_ADO_job['Total'] = df_ADO_job.groupby(['Time'])['Size'].transform('sum')
df_ADO_covid['Total'] = df_ADO_covid.groupby(['Time'])['Size'].transform('sum')
df_ADO_fam = df_ADO_fam.drop_duplicates(subset=["Time"])
df_ADO_job = df_ADO_job.drop_duplicates(subset=["Time"])
df_ADO_covid = df_ADO_covid.drop_duplicates(subset=["Time"])

df_ADO_fam = df_ADO_fam[['Time','Total']]
df_ADO_job = df_ADO_job[['Time','Total']]
df_ADO_covid = df_ADO_covid[['Time','Total']]
```

```python
df_ADO_fam.to_csv('../Data/ADO_fam.csv')
df_ADO_job.to_csv('../Data/ADO_job.csv')
df_ADO_covid.to_csv('../Data/ADO_cov.csv')
```
